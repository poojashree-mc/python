```python
import requests

# === Config ===
owner = "torvalds"
repo = "linux"
url = f"https://api.github.com/repos/{owner}/{repo}/commits"

# === Optional: Add headers for better rate limit (esp. with token) ===
headers = {
    "Accept": "application/vnd.github+json",
    # "Authorization": "Bearer YOUR_GITHUB_TOKEN"  # optional for private or large access
}

# === Fetch commits ===
response = requests.get(url, headers=headers)

if response.status_code == 200:
    commits = response.json()
    for commit in commits[:10]:  # show only latest 10
        sha = commit['sha'][:7]
        author = commit['commit']['author']['name']
        date = commit['commit']['author']['date']
        message = commit['commit']['message'].split('\n')[0]
        print(f"{sha} | {author} | {date}\n→ {message}")
        print('-' * 60)
else:
    print("Error:", response.status_code, response.text)
```
